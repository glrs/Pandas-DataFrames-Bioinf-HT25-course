### Challenge 1: Print the 'Name', 'Type 1' and 'Attack' columns of the Legendary pokemon

### Challenge 2: How many non-Legendary pokemon are there? Also list how many Pokemon per Type 1.

### Challenge 3:
#### Show the `Name`, `Type 1`, and `HP` of all **Fire-type** Pokemon with `HP > 70`, sorted by `HP` descending.

```python
fire = pkm.loc[ : , 'Type 1' ] == 'Fire'
high_hp = pkm.loc[:, 'HP'] > 70

pkm.loc[ fire & high_hp , ['Name', 'Type 1', 'HP']].sort_values(by='HP', ascending=False)

# Alternatively in an one-liner
# pkm.loc[(pkm.loc[:, "Type 1"] == "Fire") & (pkm.loc[: , "HP"] > 70), ["Name", "Type 1", "HP"]].sort_values(by="HP", ascending=False)
```

### Challenge 5: New derived column + save CSV

Create a new column `Power` defined as `Attack + Sp. Atk + Speed`.
Then show the **top 10** Pokemon by `Power` (print `Name`, `Type 1`, `Power`), and finally, save this table to `top_power.csv` (no index).

```python
pkm.loc[:, "Power"] = pkm.loc[:, 'Attack'] + pkm.loc[:, "Sp. Atk"] + pkm.loc[:, "Speed"]
```

Below there are some alternatives. Since we replace the original `pkm` dataframe, running more than one of these will probably fail. Pick one and start over, or save the results in new dataframe variables (i.e. `pkm1`, `pkm2`, etc.)

```python
# Alternatively!
# power = pkm['Attack'] + pkm['Sp. Atk'] + pkm['Speed']
# power.name = "Power"
# pkm = pd.concat([pkm, power], axis=1)
# pkm
```

```python
top_p = pkm.loc[:, ["Name", "Type 1", "Power"]].sort_values(by="Power", ascending=False).head(10)
top_p.to_csv("top_power.csv", index=False)
```
